# Get all words

```python
def get_categories(url):
    r = requests.get(url)
    soup = BeautifulSoup(r.content, 'html.parser')
    list_rows = soup.select('.card__content a')

    categories = {}
    for row in list_rows:
        categories[row.text.strip()] = [row['href'],[]]
        
    return categories
```

```python
url = 'https://www.enchantedlearning.com/wordlist/'
categories = get_categories(url)
categories
```

```python
base_url = 'https://www.enchantedlearning.com/'


already_completed = []
def get_words(url, categories, sleep):
    
    i = 0
    for key in categories:
        key_url = categories[key][0]
        word_url = base_url+key_url
        r = requests.get(word_url)
        soup = BeautifulSoup(r.content, 'html.parser')

        list_rows = soup.select('.wordlist-item')
        for row in list_rows:
            categories[key][1].append(row.text.strip())

        i=i+1
        print(i," ",key, end=" ")
        time.sleep(sleep)
    return categories


categories = get_words(base_url, categories, 5)
```

```python
for key in ['All','Nouns','Articles','Abbreviations','Prefixes','Punctuation']:
    if key in categories:
        del categories[key]

i = 0
for key in categories:
    i=i+1
print(i)
```
